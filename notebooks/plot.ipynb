```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['train', 'val', 'test']
values1 = [0.871, 0.7244, 0.754]
values2 = [0.9833, 1, 0.7857]

# Create figure and axes
fig, ax = plt.subplots()

# Width of a bar
width = 0.3

# Setting position of bar on X axis
pos = np.arange(len(categories))

# Make the plot
rects1 = ax.bar(pos - width/2, values1, width, label='Baseline', color='#CCCCCC')  # Use a lighter shade of grey
rects2 = ax.bar(pos + width/2, values2, width, label='TuneScout', color='#94C175')  # Use '#94C175' for the color

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xticks(pos)
ax.set_xticklabels(categories)
ax.legend()

# Format y-axis as percentages
ax.set_yticklabels(['{:,.0%}'.format(x) for x in ax.get_yticks()])
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Hide the top, right, and left spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Display the plot
plt.show()
```
